```python
import pandas as pd
import yfinance as yf
from yahoofinancials import YahooFinancials

import plotly.express as px
```

```python
def plot_stock_prc(stock_list):
    stock_plot = pd.DataFrame()
    for ticker in stock_list:
        # scrape stock movements from yahoo finance
        aapl_df = yf.download(ticker, 
                        start= '2019-01-01',  # start='2010-12-01', # start 1 month before
                        end = '2020-12-30',   # end='2022-01-30', # end 1 month later
                        progress=False,)
        aapl_df = aapl_df.reset_index(drop=False)
        stock_plot['Date']=aapl_df['Date']
        stock_plot[ticker]=aapl_df['Close']
    # plot line
    fig = px.line(stock_plot, x="Date", y=stock_plot.columns,
                hover_data={"Date": "|%B %d, %Y"},
                title='Closing price')
    # update axes
    fig.update_xaxes(
        dtick="M1",
        tickformat="%b\n%Y")
    return(fig)
```

```python
japan = [ '8766.T', '8630.T','8725.T']
# plot line graph
fig = plot_stock_prc(japan)
# Add shapes
fig.update_layout(
    shapes=[dict( fillcolor="rgba(63, 81, 181, 0.2)", line={"width": 0}, type="rect",
                    x0="2019-05-18", x1="2019-05-30", xref="x",
                    y0=0, y1=0.95, yref="paper"),
            dict( fillcolor="rgba(63, 81, 181, 0.2)", line={"width": 0}, type="rect",
                    x0="2019-11-15", x1="2019-11-30", xref="x",
                    y0=0, y1=0.95, yref="paper"),
            dict( fillcolor="rgba(63, 81, 181, 0.2)", line={"width": 0}, type="rect",
                    x0="2020-05-15", x1="2020-05-25", xref="x",
                    y0=0, y1=0.95, yref="paper"),
            ]
)
fig.show()
```
